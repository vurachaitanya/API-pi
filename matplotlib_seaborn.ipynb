```
import pandas
import matplotlib.pyplot as plt
import seaborn as sns

data=pandas.read_csv("https://raw.githubusercontent.com/vurachaitanya/API-pi/master/Churn_Modelling.csv")
#data.describe
#data.columns
data.drop(['RowNumber','CustomerId','Surname'],axis=1,inplace=True)

plt.figure(figsize=(12,5))
sns.countplot(data.Geography)
plt.show()
# Customer of France has good strong for the bank
```

```
sns.countplot(data.Geography[data['Exited']==1])
#in which country we have huge impact ie Germany are laving the customer are more ie 5K to 800(10:1.6) & 2.5K to 800 (10:5)
```

```
plt.figure(figsize=(12,8))
plt.subplot(211) ## 2nd row 1st column
sns.countplot(data.Geography)
plt.subplot(212) # 2nd row 2nd column
sns.countplot(data.Geography[data['Exited']==1])

```

```
plt.figure(figsize=(12,5))
sns.countplot(data['Geography'],hue=data['Exited'])
plt.show()
```

```
plt.figure(figsize=(12,8))
plt.subplot(211)
sns.countplot(data['HasCrCard'])
plt.subplot(212)
sns.countplot(data['HasCrCard'][data['Exited']==1])
plt.show() #Creid card who are having is not impacting the leaving the bank customers.
```

```
plt.figure(figsize=(12,5))
sns.factorplot(x='Geography',col='HasCrCard',data=data[data.Exited==1],kind='count')
plt.show() # Having credit card and not having is impacting different contries in different manner.
```
