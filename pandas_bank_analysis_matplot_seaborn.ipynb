```
import pandas
data=pandas.read_csv("https://raw.githubusercontent.com/vurachaitanya/API-pi/master/Churn_Modelling.csv")
#data.describe
data.columns
data.drop(['RowNumber','CustomerId','Surname'],axis=1,inplace=True)
#data=data.iloc[:,3:]
#if there is any missing values
print(data.isnull().sum())
```

```
#impact of age on people leaving or not leaving
print(data['Age'][data['Exited']==1].mean())
print(data['Age'][data['Exited']==0].mean())
print(data['Age'].describe())
```

```
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12,5))
sns.distplot(data['CreditScore'][data['Exited']==1])
sns.distplot(data['CreditScore'][data['Exited']==0])
plt.legend(['1','0'])
plt.show()
#Credit Score is not impacting the bank who are leaving.
```

```
import matplotlib.pyplot as plt
import seaborn as sns # As sns represents using alias to use its short form
plt.figure(figsize=(12,5))
sns.distplot(data['Age'][data['Exited']==1])
sns.distplot(data['Age'][data['Exited']==0])
plt.legend(['1','0'])
plt.show()
#Age after 45 years people are leaving so bank should offer better offers to old age who are more than 45 years.
```

# Matplotlib 
* offers line, scatter, histogram, pi charts, boxplots
* Functionalities to customize and create and show the figure window.
# Seaborn
* offers distribution plots, violin plots, heatmaps, boxplots,clusetmaps,regression plots, factor plots

```
import numpy
import matplotlib.pyplot as plt
import seaborn as sns

x=numpy.arange(0,10)
y=numpy.sin(x)
z=numpy.cos(x)

plt.figure(figsize=(12,4))
plt.plot(x,y,'r',label='sin[x]')
plt.plot(x,z,'g',label='cos[x]')
plt.legend(['1','0'])
#plt.plot(x,y)
plt.show()
```

```
plt.figure(figsize=(12,4))
plt.scatter(x,y,label='sin[x]')
plt.scatter(x,z,label='cos[x]')
plt.legend(['1','0'])
plt.show()
```

```
import numpy
import matplotlib.pyplot as plt
import seaborn as sns

x=numpy.arange(0,10)
y=numpy.sin(x)
z=numpy.cos(x)

plt.figure(figsize=(12,4))
plt.scatter(x,y,label='sin[x]')
plt.scatter(x,z,label='cos[x]')
plt.legend(['1','0'])
plt.title('Grapth of cosx & sinx')
plt.xlabel('value of x')
plt.ylabel('value of cosx & Sinx')
plt.show()
```
